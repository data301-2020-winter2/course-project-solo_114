**Dataset**

- The column 'instant' refers to the day that the information was collected for. 1 represents the first day of collection (Jan. 1, 2011), 2 represents the second day of collects (Jan. 2, 2011), and so on.

- The column 'weathersit' refers to the weather conditions observed each day and is on a scale of 1-4. 1 represents clear skies, 2 represents clouds and mist, 3 represents light rain and snow, and 4 represents rain and storms.

- The column 'casual' refers to the count of the casual bike users on the observed day. Casual users are those who have not registered with the system, and rather just use the system whenever they need to.

- The column 'registered' refers to the count of registered bike users on the observed day. Registered users are those who have registered with the system and are therefore 'members'.

- The column 'count' refers to the total count of users, both count and registerd, that were observed each day.

**Analysis**
   The bike sharing dataset was provided by the Center for Machine Learning and Intelligent Systems, and was originally collected by the Laboratory of Artificial Intelligence and Decision Support from the University of Porto. This dataset provides information about different factors that affect the frequency of bikes being rented in the bike sharing system. These factors include the weather situation - such as rainy, stormy, and sunny -, count, casual bike users, and registered bike users. In this analysis, we are curious about learning the different rates of casual and registered bike users in order to better assess how the system is used. Additionally, we are also curious to know how the varying weather situations might affect counts of both casual and registered bike users in order to know what weather conditions, and therefore locations, bike sharing systems like this one might work best in.
    
   The bike sharing system was put into place in Washington D.C., and shows different counts of people everyday using this system. Some of these people registered with a bike sharing system, while others are only casual users. Upon analyzing the dataset, there appear to be over four times as many registered users as there are casual. As observed in Table 1, on average, there are only around 848 casual users per day, while for registered users the count is 3,656. This may be because registered users have committed to registering to the system, and are therefore likely to be using bikes more often than casual users, who tend to rent the bikes on occasion, in order to get the full value and experience of having registered. These averages come from great ranges of users, with the minimum number of casual users in one day being only 2, and 22 for registered users. The maximum count for casual users was 3,410 and was 6,946 for registered users.

   When visualizing the frequency of different counts of registered and casual users, we are able to identify more patterns. The histogram of casual users shows a right skewed distribution (Figure 1). There are many days where there are lower rates of bike rentals. As the amount of people using bikes per day increases the frequency decreases. This aligns with the evidence stated before, with casual users not being as committed to using the system everyday, but rather rent the bikes when they see fit. In comparison, the histogram of registered users shows a relatively even distribution (Figure 2).  There are few days when there are either many people using the bikes or very few people using the bikes; however,  there are commonly days where the amount of people using the bikes is around 3,000 to 5,000. These rates are still fairly high, especially when compared to those of casual users, showing that even without a high count of the highest number of rentals, there are still a very good position of registered users who use the system everyday.
    
   Additionally, Figure 3 and Figure 4 show that despite different counts of users between casual and registered, both people who have registered and people who are casual users tend to follow similar patterns of use. Both show high counts of rentals around instant 150-250 and 500-650, while showing lower counts between 300-400. Days 150-250 and 500-650 roughly correspond with the summer, while days 300-400 roudlghy correspond to the winter. This makes sense as it is more desirable to rent and ride bikes during the summer rather than the winter. It also indicates that while the difference between being a registered or casual user has a large effect on the frequency of users, there are other factors at play that are affecting the distribution of frequencies throughout the observed days.
    
   Upon delving deeper into the dataset, there are many attributes about environmental conditions for each day, most notably, the weather situation. The weather situation is ranked on a scale of 1 to 4. 1 represents days where there is a clear sky or it is partly cloudy. 2 represents days with mist and clouds. 3 represents days will light snow, light rain and thunderstorms or scattered clouds. Finally, 4 represents days with heavy rain and ice pellets, thunderstorms and mist, and snow and fog. As one can imagine, days where the weather situation falls into Category 3 or 4 do not seem like the best days to be using a bike as a means of transportation as the weather is more severe, and for both casual and registered users, this held true. While there were no observations of category 4, on days when the weather situation was in Category 3, both casual and registered users saw significantly lower counts of people using a bike share system. Curiously, despite the mutual decrease in counts on days with weather falling in Category 3, there was a much greater range of registered users on these days (Figure 5 and 6). This suggests that some registered users are more likely to put up with the weather and continue to rent the bikes than casual users.
    
   In conclusion, the bike sharing dataset from Washington, D.C. shows some interesting patterns about the frequency of bike rentals of casual versus registered users, as well as how they are affected by different weather situations. Registered users showed greater frequencies of users than casual users, suggesting that when people sign up to register for the system, they are likely to use it more frequently as they have committed to renting bikes. Additionally, when comparing the counts of rentals with the weather situation, we found that as the weather is more extreme, the count of both registered and casual users decrease.
