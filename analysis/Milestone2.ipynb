1. The bike sharing system was put into place in Washington D.C., and from January 1, 2011 to December 31, 2012 shows different counts of people everyday using this system. Some of these people registered with the bike sharing system, while others are only casual users. Upon analyzing the data set, there appear to be over four times as many registered users as there are casual. On average, there are only around 848 casual users per day, while for registered users the count is 3656. This may be because registered users have committed to registering to the network, and are therefore likely to be using bikes more often than casual users, who tend to use the bikes on occasion. These averages come from great ranges of users, with the minimum number of casual users in one day being only 2 and 22 for registered users, while the maximum count for casual users was 3410 and was 6946 for registered users. These large ranges show that there are more factors than just whether a person is registered with the network that affect the usage of the bike sharing system. 
2. When visualizing the frequency of different counts of registered and casual users, we are able to identify more patterns. The histogram of casual users shows a right skewed distribution. There are many days where there are lower rates of bike use. As the amount of people using bikes per day increases the frequency decreases. This further shows that casual users are much more likely to be using the bikes on occasion rather than using it as an everyday means of transport. In comparison, the histogram of registered users shows an even distribution. There are few days when there are either many people using the bikes or very few people using the bikes; however, there are commonly days where the amount of people using the bikes is around 3000 to 5000. 
3. Upon delving deeper into the data set, there are many attributes about environmental conditions for each day, most notably, the weather situation. The weather situation is ranked on a scale of 1 to 4. 1 represents days where there is a clear sky or it is partly cloudy. 2 represents days with mist and clouds. 3 represents days will light snow, light rain and thunderstorms or scattered clouds. Finally, 4 represents days with heavy rain and ice pellets, thunderstorms and mist, and snow and fog. As one can imagine, days where the weather situation falls into Category 3 or 4 do not seem like the best days to be using a bike as a means of transportation, and for both casual and registered users, this held true. While there was no category 4 observed, on days when the weather situation was in Category 3, both casual and registered users saw significantly lower counts of people using the bike sharing system. Curiously, there were many more outliers to this pattern for registered users than there were for casual users, suggesting that some registered users are more likely to put up with the weather and continue to use the bikes. In addition, casual and registered users showed the greatest counts of users when the weather situation fell into category 1. This is likely because category 1 represents much nicer weather where people are more likely to want to be outside and are more willing to use bikes as a means of transportation.
