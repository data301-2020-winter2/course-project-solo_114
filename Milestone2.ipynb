```python
import pandas as pd
import numpy as np

data = pd.read_csv('data/rawdata/Bike-Sharing-Dataset/day.csv')

df = (pd.DataFrame(data=data)
       .rename(columns={"atemp":"air_temp","cnt":"count"})
       .dropna()
       .drop(columns=["dteday","season","yr","mnth","weekday","workingday","holiday"]))
df.head()
```

```python
def load_and_process(data):
    df1 = (pd.DataFrame(data=data)
       .rename(columns={"atemp":"air_temp","cnt":"count"})
       .dropna())
    df2 = (df1
       .drop(columns=["dteday","season","yr","mnth","weekday","workingday","holiday"]))
    return df2

load_and_process(data)
```
